```python
from sympy.plotting import plot_parametric as plt_p   # 媒介変数表示プロット
from matplotlib import rc
rc('text', usetex=True)

from sympy import assoc_legendre
from sympy import factorial, sqrt
from sympy import sin, cos, pi
from sympy.plotting import plot
from sympy import Symbol

sympy.init_printing()
# いい感じに出力するよう設定してくれるらしい

t = Symbol('t')  # 位相変数

def Pnm(n, m, t):
    return (-1)**m * assoc_legendre(n, m, cos(t)) * sqrt((2*n+1)*factorial(n-m)/2/factorial(n+m))

def x(n, m, t):
    return abs(Pnm(n, m, t)) * cos(t)

def y(n, m, t):
    return abs(Pnm(n, m, t)) * sin(t)


p = plt_p((x(4, 0, t), y(4, 0, t)), (x(4, 1, t), y(4, 1, t)),
                  (x(4, 2, t), y(4, 2, t)), (x(4, 3, t), y(4, 3, t)),
                  (x(4, 4, t), y(4, 4, t)),
                 (t, 0, 2*pi), legend=True, show=False)
                 # xlim=(-1.8, 1.8), 
                 # xlabel = '$x$', ylabel='')

p[0].line_color = '#FF6860'
p[0].label = r'$|\bar{P}_4^0 (\cos\theta)|$'
p[1].line_color = '#4169E1'
p[1].label = r'$|\bar{P}_4^1 (\cos\theta)|$'
p[2].line_color = '#FDC44F'
p[2].label = r'$|\bar{P}_4^2 (\cos\theta)|$'
p[3].line_color = '#7CBE89'
p[3].label = r'$|\bar{P}_4^3 (\cos\theta)|$'
p[4].line_color = '#F6ADC6'
p[4].label = r'$|\bar{P}_4^4 (\cos\theta)|$'
# p[5].line_color = '#00DDFF'
# p[5].label = '$P_5 (x)$'

p.show()

p.save('assoc_legendre_cos4.pdf')

```

```python
from sympy.plotting import plot_parametric as plt_p   # 媒介変数表示プロット
from matplotlib import rc
rc('text', usetex=True)

from sympy import legendre
from sympy import factorial, sqrt
from sympy import sin, cos, pi
from sympy.plotting import plot
from sympy import Symbol

sympy.init_printing()
# いい感じに出力するよう設定してくれるらしい

t = Symbol('t')  # 位相変数

def Pn(n, t):
    return  legendre(n, cos(t))

def x(n, t):
    return abs(Pn(n, t)) * cos(t)

def y(n, t):
    return abs(Pn(n, t)) * sin(t)


p = plt_p((x(0, t), y(0, t)), (x(1, t), y(1, t)),
                  (x(2, t), y(2, t)), (x(3, t), y(3, t)),
                  (x(4, t), y(4, t)), 
                 (t, 0, 2*pi), legend=True, show=False,
                 xlim=(-1.6, 1.6))
                 # xlabel = '$x$', ylabel='')

p[0].line_color = '#FF6860'
p[0].label = r'$|P_0 (\cos\theta)|$'
p[1].line_color = '#4169E1'
p[1].label = r'$|P_1 (\cos\theta)|$'
p[2].line_color = '#FDC44F'
p[2].label = r'$|P_2 (\cos\theta)|$'
p[3].line_color = '#7CBE89'
p[3].label = r'$|P_3 (\cos\theta)|$'
p[4].line_color = '#F6ADC6'
p[4].label = r'$|P_4 (\cos\theta)|$'

p.show()

p.save('legendre_cos.pdf')
```
