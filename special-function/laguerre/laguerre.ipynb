```python
import sympy as sym
z = sym.Symbol('z')


# Associated Laguerre polynomials
def L(n, k):
    f = (sym.exp(-z) * z**n) 
    
    for i in range(n-k):
        f = sym.diff(f)
    
    return sym.exp(z) * z**(-k) * f * sym.Rational(sym.factorial(n) / sym.factorial(n-k)) * (-1)**k


# print
for n in range(7):
    for k in range(n+2):
        print('L_' + str(n) + '^' + str(k) + ' (z) =', sym.factor(sym.expand(L(n, k))))
    print('')
```

```python
# print
for n in range(7):
    for k in range(n+2):
        V = sym.factorial(n)**2 / sym.factorial(n-k) / sym.factorial(k) * (-1)**k
        print('L_' + str(n) + '^' + str(k) + ' (0) =', V)
    print('')
```

```python
# plot
# title

from matplotlib import rc
rc('text', usetex=True)

import sympy
from sympy import factorial, sqrt
from sympy import pi
from sympy.plotting import plot
from sympy import Symbol

sympy.init_printing()
# いい感じに出力するよう設定してくれるらしい


# 規格化Laguerre陪多項式
def L_(n, m):
    return L(n, m)   / factorial(n)

p = plot(L_(0, 0), L_(1, 0), L_(2, 0), L_(3, 0),  L_(4, 0),
                (z, -2, 12), 
                 legend=True, show=False, 
                 ylim=(-10, 20),
                 xlabel = r'$x$', ylabel='')

p[0].line_color = '#FF6860'
p[0].label = r'$\bar{L}_0 (x)$'
p[1].line_color = '#4169E1'
p[1].label = r'$\bar{L}_1 (x)$'
p[2].line_color = '#FDC44F'
p[2].label = r'$\bar{L}_2 (x)$'
p[3].line_color = '#7CBE89'
p[3].label = r'$\bar{L}_3 (x)$'
p[4].line_color = '#F6ADC6'
p[4].label = r'$\bar{L}_4 (x)$'
# p[5].line_color = '#00DDFF'
# p[5].label = r'$\bar{L}_5^0 (x)$'

p.show()

p.save('laguerre_title.pdf')
```
