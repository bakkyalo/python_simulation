```python
from sympy.polys.orthopolys import spherical_bessel_fn as j
import sympy

from matplotlib import rc
rc('text', usetex=True)
sympy.init_printing()

from sympy import besselj
from sympy.plotting import plot
from sympy import Symbol
from sympy import sqrt, pi

x = Symbol('x')

# 球ベッセル関数
def j(n, x):
    return sqrt(pi / (2*x)) * besselj(n+1/2, x)


p = plot(j(0, x), j(1, x), j(2, x), j(3, x), 
                 j(4, x), j(5, x), j(10, x),
                 (x, -0.2, 15), 
                 ylim = (-0.3, 1),
                 legend=True, show=False, xlabel = '$x$', ylabel='')
p[0].line_color = '#FF6860'
p[0].label = r'$j_0 (x)$'
p[1].line_color = '#4169E1'
p[1].label = r'$j_1 (x)$'
p[2].line_color = '#FDC44F'
p[2].label = r'$j_2 (x)$'
p[3].line_color = '#58BE89'
p[3].label = r'$j_3 (x)$'
p[4].line_color = '#FF99FF'
p[4].label = r'$j_4 (x)$'
p[5].line_color = '#00DDFF'
p[5].label = r'$j_5 (x)$'
p[6].line_color = '#b8d200'
p[6].label = r'$j_{10} (x)$'

p.show()

p.save('sph_j_n.pdf')
```

```python
import sympy
from sympy import besselj as J
from sympy import bessely as N


from matplotlib import rc
rc('text', usetex=True)
sympy.init_printing()

from sympy import bessely
from sympy.plotting import plot_parametric as pp
from sympy import Symbol
from sympy import sqrt, pi
# plt.style.use('seaborn-darkgrid')
t = Symbol('t')

p = pp((J(0, t), N(0, t)), (J(1, t), N(1, t)), (J(2, t), N(2, t)), (J(3, t), N(3, t)), 
           (t, 0, 25), xlim=(-1.3, 1.3), ylim=(-1, 1), xlabel = r'$x$', ylabel=r'$y$', 
              legend =True, show=False)

p[0].line_color = '#FF6860'
p[0].label = r'$\left(J_0 (t), N_0(t)\right)$'
p[1].line_color = '#4169E1'
p[1].label = r'$\left(J_1 (t), N_1(t)\right)$'
p[2].line_color = '#FDC44F'
p[2].label = r'$\left(J_2 (t), N_2(t)\right)$'
p[3].line_color = '#58BE89'
p[3].label = r'$\left(J_3 (t), N_3(t)\right)$'

p.show()

p.save('parametric-bessel.pdf')
```
