```python
from sympy import hermite
from sympy.plotting import plot
from sympy import Symbol

x = Symbol('x')

p = plot(hermite(0, x), hermite(1, x), hermite(2, x), hermite(3, x), hermite(4, x),
                 (x, -2, 2), legend=True, show=False, xlabel = '$x$', ylabel='')
p[0].line_color = '#FF6860'
p[0].label = '$H_0 (x)$'
p[1].line_color = '#4169E1'
p[1].label = '$H_1 (x)$'
p[2].line_color = '#FDC44F'
p[2].label = '$H_2 (x)$'
p[3].line_color = '#58BE89'
p[3].label = '$H_3 (x)$'
p[4].line_color = '#C088FF'
p[4].label = '$H_4 (x)$'

p.show()

p.save('hermite.pdf')
```

```python
for i in range(0, 6):
    print(hermite(i, x))
```
