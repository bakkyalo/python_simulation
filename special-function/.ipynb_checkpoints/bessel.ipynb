```python
from sympy import besselj
from sympy.plotting import plot
from sympy import Symbol

x = Symbol('x')

p = plot(besselj(0, x), besselj(1, x), besselj(2, x), besselj(3, x),
                 (x, 0, 15), legend=True, show=False, xlabel = '$x$', ylabel='')
p[0].line_color = '#FF6860'
p[0].label = '$J_0 (x)$'
p[1].line_color = '#33BFDB'
p[1].label = '$J_1 (x)$'
p[2].line_color = '#FDC44F'
p[2].label = '$J_2 (x)$'
p[3].line_color = '#58BE89'
p[3].label = '$J_3 (x)$'

p.show()
```

```python
from sympy import besselj
from sympy.plotting import plot
from sympy import Symbol

x = Symbol('x')

p = plot(besselj(0, x), besselj(1/2, x), besselj(1, x), besselj(3/2, x), besselj(10, x),
                 (x, -15, 15), legend=True, show=False, xlabel = '$x$', ylabel='')
p[0].line_color = '#FF6860'
p[0].label = '$J_0 (x)$'
p[1].line_color = '#33BFDB'
p[1].label = '$J_{1/2} (x)$'
p[2].line_color = '#FDC44F'
p[2].label = '$J_1 (x)$'
p[3].line_color = '#58BE89'
p[3].label = '$J_{3/2} (x)$'
p[4].line_color = '#FF99FF'
p[4].label = '$J_{10} (x)$'

p.show()
```

```python
from sympy import besselj
from sympy.plotting import plot
from sympy import Symbol

x = Symbol('x')

p = plot(besselj(0, x), besselj(-1, x), besselj(-2, x), besselj(-3, x),
                 (x, -10, 10),  legend=True, show=False, xlabel = '$x$', ylabel='')
p[0].line_color = '#FF6860'
p[0].label = '$J_0 (x)$'
p[1].line_color = '#33BFDB'
p[1].label = '$J_{-1} (x)$'
p[2].line_color = '#FDC44F'
p[2].label = '$J_{-2} (x)$'
p[3].line_color = '#58BE89'
p[3].label = '$J_{-3} (x)$'

p.show()
```

```python
from sympy import bessely
from sympy.plotting import plot
from sympy import Symbol

x = Symbol('x')

p = plot(bessely(0, x), bessely(1, x), bessely(2, x), bessely(3, x),
                 (x, -5, 20) , ylim=(-1, 1), legend=True, show=False, xlabel = '$x$', ylabel='')
p[0].line_color = '#FF6860'
p[0].label = '$N_0 (x)$'
p[1].line_color = '#33BFDB'
p[1].label = '$N_1 (x)$'
p[2].line_color = '#FDC44F'
p[2].label = '$N_2 (x)$'
p[3].line_color = '#58BE89'
p[3].label = '$N_3 (x)$'

p.show()
```
