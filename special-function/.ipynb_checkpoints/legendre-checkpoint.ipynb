```python
from sympy import legendre
from sympy.plotting import plot
from sympy import Symbol

x = Symbol('x')

p = plot(legendre(0, x), legendre(1, x), legendre(2, x), legendre(3, x), 
                 legendre(4, x), legendre(5, x),
                 (x, -1, 1), legend=True, show=False, 
                 ylim=(-1.2,1.2), xlim=(-1.8, 1.8), 
                 xlabel = '$x$', ylabel='')

p[0].line_color = '#FF6860'
p[0].label = '$P_0 (x)$'
p[1].line_color = '#4169E1'
p[1].label = '$P_1 (x)$'
p[2].line_color = '#FDC44F'
p[2].label = '$P_2 (x)$'
p[3].line_color = '#7CBE89'
p[3].label = '$P_3 (x)$'
p[4].line_color = '#F6ADC6'
p[4].label = '$P_4 (x)$'
p[5].line_color = '#00DDFF'
p[5].label = '$P_5 (x)$'

p.show()

p.save('legendre.pdf')
```
