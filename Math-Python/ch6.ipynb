```python
'''
Example of using matplotlib's Circle patch
'''
import matplotlib.pyplot as plt

def create_circle():
    circle = plt.Circle((0, 0), radius = 0.5, fc = 'b', ec='g')
    return circle

def show_shape(patch):
    ax = plt.gca()
    # ax.set_aspect('equal')
    ax.add_patch(patch)
    plt.axis('scaled')
    plt.show()
    
if __name__ == '__main__':
    c = create_circle()
    show_shape(c)
    
```

```python
'''
円運動やってみる
'''
%matplotlib nbagg

import matplotlib.pyplot as plt
from matplotlib import animation
import math


def create_dot():
    x = math.cos(0)
    y = math.sin(0)
    dot = plt.Circle((x, y), radius = 0.05)
    return dot

def update_dot(i, dot):
    theta = i *  math.pi * 0.02
    x = math.cos(theta)
    y = math.sin(theta)
    dot.center = x, y
    return dot,

fig = plt.gcf()
ax = plt.axes(xlim=(-1.5, 1.5), ylim=(-1.5, 1.5))
ax.set_aspect('equal')
ax.grid()

dot = create_dot()
ax.add_patch(dot)

anim = animation.FuncAnimation(
    fig,
    update_dot,
    fargs = (dot,),
    frames = 100,
    interval = 3,
    repeat = True
    )
plt.title('circulate motion')
plt.show()
```

```python
'''
Example of selecting a transformation from two equally probable transformations
線上の酔歩(ランダムウォーク)
'''
import matplotlib.pyplot as plt
import random

def transformation_1(p):
    x = p[0]
    y = p[1]
    return x + 1, y - 1

def transformation_2(p):
    x = p[0]
    y = p[1]
    return x + 1, y + 1

def transform(p):
    # 変数関数のリスト
    transformations = [transformation_1, transformation_2]
    # pick a random transformation function and call it
    t = random.choice(transformations)
    x, y = t(p)    # tはtransformations という関数の配列から選んだ関数。これをpに作用させる
    return x, y

def build_trajectory(p, n):
    x = [p[0]]
    y = [p[1]]
    for i in range(n):
        p = transform(p)
        x.append(p[0])
        y.append(p[1])
    return x, y

if __name__ == '__main__':
    # initial point
    p = (1, 1)
    n = int(input('Enter the number of iterations: '))
    x, y = build_trajectory(p, n)
    
    # plot
    plt.plot(x, y)
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.show()
```

```python
'''
Draw Barnsley Fern
バーンスレイのシダ
'''
%matplotlib nbagg
import random
import matplotlib.pyplot as plt

def transformation_1(p):
    x = p[0]
    y = p[1]
    x1 = 0.85*x + 0.04*y
    y1 = -0.04*x + 0.85*y + 1.6
    return x1, y1

def transformation_2(p):
    x = p[0]
    y = p[1]
    x1 = 0.2*x - 0.26*y
    y1 = 0.23*x + 0.22*y + 1.6
    return x1, y1

def transformation_3(p):
    x = p[0]
    y = p[1]
    x1 = -0.15*x + 0.28*y
    y1 = 0.26*x + 0.24*y + 0.44
    return x1, y1

def transformation_4(p):
    x = p[0]
    y = p[1]
    x1 = 0
    y1 = 0.16*y
    return x1, y1

def get_index(probability):    # probability = [0.85, 0.07, 0.07, 0.01]
    r = random.random()
    c_probability = 0
    sum_probability = []
    for p in probability:
        c_probability += p
        sum_probability.append(c_probability)
    for item, sp in enumerate(sum_probability):    # enumerate で (インデックス, 要素) の順に取得できる
        if r <= sp:
            return item
    return len(probability) - 1

def transform(p):
    # 変換関数のリスト
    transformations = [transformation_1, transformation_2, transformation_3, transformation_4]
    probability = [0.85, 0.07, 0.07, 0.01]
    # pick a random transformation function and call it
    tindex = get_index(probability)
    t = transformations[tindex]
    x, y = t(p)
    return x, y

def draw_fern(n):
    # We start with (0, 0)
    x = [0]
    y = [0]
    
    x1, y1 = 0, 0
    for i in range(n):
        x1, y1 = transform((x1, y1))
        x.append(x1)
        y.append(y1)
    return x, y

if __name__ == '__main__':
    n = int(input('Enter the number of points in the Fern: '))
    x, y = draw_fern(n)
    # plot the points
    plt.plot(x, y, 'go', markersize = 3)
    plt.title('Fern with {0} poiints'. format(n))
    plt.show()
```

```python
# imshow関数
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import random
import time
% matplotlib nbagg

start = time.time()

def initialize_image(x_p, y_p):
    image = []
    for i in range(y_p):
        x_colors = []
        for j in range(x_p):
            x_colors.append(0)
        image.append(x_colors)
    return image

def color_points():
    x_p = 20
    y_p = 20
    image = initialize_image(x_p, y_p)
    for i in range(y_p):
        for j in range(x_p):
            image[i][j] = random.randint(0, 10)
    plt.imshow(
        image, 
        origin = 'lower', 
        extent = (0, 5, 0, 5),
        cmap = cm.Greys_r, 
        interpolation = 'nearest'
    )
    plt.colorbar()
    plt.show()

if __name__ == '__main__':
    color_points()
    
end = time.time()

print('{0} 秒\n'. format(end - start))
```

```python
# マンデルブロ集合
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import random
% matplotlib nbagg

def color_points():
    x_p = 600
    y_p = 600
    max_iteration = 1200
    image = [[0 for i in range(x_p)] for j in range(y_p)]    # 色
    x = [[0 for i in range(x_p)] for j in range(y_p)]    # x座標
    y = [[0 for i in range(x_p)] for j in range(y_p)]    # y座標
    
    for i in range(y_p):
        for j in range(x_p):
            x[i][j] = -2.5 + j * (1.0-(-2.5)) / (x_p-1)
            y[i][j] = -1.0 + i * (1.0-(-1.0)) / (y_p-1)
    
    for k in range(y_p):
        for i in range(x_p):
            z1 = 0 + 0j
            c = x[k][i] + y[k][i] * 1j
            iteration = 0
            
            for iteration in range(max_iteration):
                z1 = z1**2 + c
                if abs(z1) >= 2:
                    break
            
            image[k][i] = iteration

    plt.imshow(
        image, 
        origin = 'lower', 
        extent = (-2.5, 1.0, -1.0, 1.0),    # (xmin, xmax, ymin, ymax)
        cmap = cm.Greys_r, 
        interpolation = 'nearest'
    )
    
    plt.colorbar()
    plt.show()

if __name__ == '__main__':
    color_points()
```
