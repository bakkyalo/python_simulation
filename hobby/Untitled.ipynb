```python
import sympy as sym
from sympy.plotting import plot

x = sym.Symbol('x')

def H(n):
    F = sym.exp(-x**2)
    for i in range(n):
        F = sym.diff(F)
        
    return (-1)**n * sym.exp(x**2) * F

H0 = H(0)
H1 = H(1)
H2 = H(2)
H3 = H(3)


for i in range(5):
    print('H_'+ str(i) + '(x) = ' + str(sym.expand(H(i))))
    
# plot
plot(H0, H1, H2, H3, (x, -25, 25))
sym.plotting.plot(H(1), (x, -25, 25))
```
