```python
%matplotlib nbagg
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation
from mpl_toolkits.mplot3d import Axes3D

k1 = 1
k2 = 2
w = 2*np.pi/10

def wave(x, y, t):
    return np.sin(k1*x + k2*y - w*t)

def update(t, Z, line):
    Z = wave(X, Y, t)
    ax.clear()
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('wave')
    line = ax.plot_surface(X, Y, wave(X, Y, t))
    ax.set_zlim3d(bottom=-3, top=3)
    return line

fig = plt.figure()
ax = Axes3D(fig)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('wave')

x = np.arange(-1.5*np.pi, 1.5*np.pi, 0.2)
y = np.arange(-1.5*np.pi, 1.5*np.pi, 0.2)
X, Y = np.meshgrid(x, y)
Z = wave(X, Y, 0)
line = ax.plot_surface(X, Y, Z)
ax.set_zlim3d(bottom=-3, top=3)

ani = animation.FuncAnimation(
    fig,
    update,
    fargs = (Z, line),
    interval = 100,
    frames = 20,
    repeat = False,
    blit = True
)

# ani.save('wave.gif', writer = 'imagemagick')
plt.show()
```
