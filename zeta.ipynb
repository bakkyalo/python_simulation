```python
import mpmath as mp
import matplotlib.pyplot as plt
from matplotlib import rc
rc('text', usetex=True)

xmin = -25
xmax = 25
ymin = -0.1
ymax = 0.1
N = 5000

x = []
zeta = []
for i in range(N+1):
    point = xmin + (xmax-xmin)*i / N
    x.append(point)
    if point == 1:
        zeta.append(float('nan'))
    else:
        zeta.append(float(mp.zeta(point)))

ax = plt.subplot()
ax.plot(x, zeta)
ax.set_xlabel(r'$x$')
ax.set_ylabel(r'$\zeta (x)$')
ax.set_ylim([ymin, ymax])
ax.grid()

# plt.savefig('zeta_minus.png')
plt.show()
```
